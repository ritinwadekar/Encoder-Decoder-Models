```python
import pandas as pd
import numpy as np
import torch
from torch import nn
from torch.optim import AdamW
from transformers import AlbertTokenizer, AlbertModel
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import f1_score, precision_score, recall_score
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
import os
import wandb
from huggingface_hub import login as hf_login
from google.colab import userdata

# First, mount Google Drive with explicit output
drive.mount('/content/drive', force_remount=True)
print("Google Drive mounted at /content/drive")

# List files in your Drive root to verify access
print("Files in Google Drive root:")
for file in os.listdir('/content/drive/MyDrive'):
    print(f" - {file}")

# Setup API keys securely using Google Colab secrets
try:
    # Try to get Wandb API key from Colab secrets
    wandb_key = userdata.get('')
    if wandb_key:
        print("Found wandb API key in Colab secrets")
        wandb.login(key=wandb_key)
    else:
        # If not in secrets, you might need to login interactively
        print("No wandb API key found in Colab secrets, login interactively")
        wandb.login()

    # Try to get Hugging Face token from Colab secrets
    hf_token = userdata.get('')
    if hf_token:
        print("Found Hugging Face token in Colab secrets")
        hf_login(token=hf_token)
    else:
        print("No Hugging Face token found in Colab secrets")
except Exception as e:
    print(f"Error setting up API keys: {str(e)}")
    print("Continuing without proper authentication...")

# Custom Dataset class with fixed tensor conversion
class EmotionDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=128):
        self.tokenizer = tokenizer
        self.data = dataframe
        self.text = dataframe['Tweet'].values
        self.max_length = max_length
        # Get emotion labels if they exist
        self.emotion_cols = ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'love',
                           'optimism', 'pessimism', 'sadness', 'surprise', 'trust']

        if all(col in dataframe.columns for col in self.emotion_cols):
            # Fix: First convert any string values to NaN, then to float32
            emotion_data = dataframe[self.emotion_cols].replace(['NONE', 'None', 'none'], np.nan)
            emotion_data = emotion_data.apply(pd.to_numeric, errors='coerce')
            self.targets = emotion_data.fillna(0).values.astype(np.float32)
        else:
            self.targets = None

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        text = str(self.text[index])

        # Tokenize the text
        inputs = self.tokenizer.encode_plus(
            text,
            None,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_token_type_ids=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        item = {
            'input_ids': inputs['input_ids'].flatten(),
            'attention_mask': inputs['attention_mask'].flatten(),
            'token_type_ids': inputs['token_type_ids'].flatten(),
        }

        if self.targets is not None:
            item['targets'] = torch.tensor(self.targets[index], dtype=torch.float)

        return item

# Model Architecture
class EmotionClassifier(nn.Module):
    def __init__(self, n_classes=11, dropout_rate=0.2):
        super(EmotionClassifier, self).__init__()
        self.albert = AlbertModel.from_pretrained('albert-base-v2')
        self.drop = nn.Dropout(p=dropout_rate)
        self.out = nn.Linear(self.albert.config.hidden_size, n_classes)

    def forward(self, input_ids, attention_mask, token_type_ids):
        outputs = self.albert(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids
        )

        # Get the [CLS] token representation
        pooled_output = outputs.pooler_output
        output = self.drop(pooled_output)
        return self.out(output)

# Fixed function to calculate F1 score during training
def calculate_f1_score(y_true, y_pred):
    """
    Calculate macro F1 score for multi-label classification

    Args:
        y_true: numpy array of true labels
        y_pred: numpy array of predicted labels (after thresholding)

    Returns:
        macro_f1: macro-averaged F1 score
    """
    # Make sure inputs are numpy arrays
    if isinstance(y_true, torch.Tensor):
        y_true = y_true.cpu().numpy()
    if isinstance(y_pred, torch.Tensor):
        y_pred = y_pred.cpu().numpy()

    # Calculate F1 score for each class
    f1_scores = []
    for i in range(y_true.shape[1]):
        # Skip classes with no positive examples in this batch
        if np.sum(y_true[:, i]) > 0:
            class_f1 = f1_score(y_true[:, i], y_pred[:, i], zero_division=0)
            f1_scores.append(class_f1)
    # Return 0 if no valid classes
    if len(f1_scores) == 0:
        return 0.0

    # Calculate macro F1 score
    return np.mean(f1_scores)

# Updated training function with wandb integration
def train_model(model, train_loader, val_loader, epochs=15, lr=2e-5, model_save_path='best_albert_model.pt'):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    model.to(device)

    # Binary Cross Entropy with Logits Loss for multi-label classification
    criterion = nn.BCEWithLogitsLoss()

    # Optimizer with weight decay
    optimizer = AdamW(model.parameters(), lr=lr, weight_decay=1e-4)

    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.5, patience=1, verbose=True
    )

    best_f1 = 0
    history = {'train_loss': [], 'val_loss': [], 'val_f1': []}

    for epoch in range(epochs):
        print(f"\nEpoch {epoch+1}/{epochs}")

        # Training phase
        model.train()
        train_loss = 0
        all_train_preds = []
        all_train_targets = []

        for batch in train_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            targets = batch['targets'].to(device)

            optimizer.zero_grad()

            outputs = model(input_ids, attention_mask, token_type_ids)
            loss = criterion(outputs, targets)

            loss.backward()
            optimizer.step()

            train_loss += loss.item()

            # Store predictions and targets for F1 calculation
            preds = torch.sigmoid(outputs) > 0.5
            all_train_preds.append(preds.cpu().detach().numpy())
            all_train_targets.append(targets.cpu().numpy())

        train_loss = train_loss / len(train_loader)
        history['train_loss'].append(train_loss)
        # Calculate training F1 score
        all_train_preds = np.vstack(all_train_preds)
        all_train_targets = np.vstack(all_train_targets)
        train_f1 = calculate_f1_score(all_train_targets, all_train_preds)

        # Validation phase
        model.eval()
        val_loss = 0
        all_val_preds = []
        all_val_targets = []

        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                token_type_ids = batch['token_type_ids'].to(device)
                targets = batch['targets'].to(device)

                outputs = model(input_ids, attention_mask, token_type_ids)
                loss = criterion(outputs, targets)

                val_loss += loss.item()

                # Apply sigmoid and threshold to get binary predictions
                preds = torch.sigmoid(outputs) > 0.5
                all_val_preds.append(preds.cpu().detach().numpy())
                all_val_targets.append(targets.cpu().numpy())

        val_loss = val_loss / len(val_loader)
        history['val_loss'].append(val_loss)

        # Concatenate predictions and targets
        all_val_preds = np.vstack(all_val_preds)
        all_val_targets = np.vstack(all_val_targets)

        # Calculate macro F1 score
        val_f1 = calculate_f1_score(all_val_targets, all_val_preds)
        history['val_f1'].append(val_f1)

        print(f"Train Loss: {train_loss:.4f}, Train F1: {train_f1:.4f}, Val Loss: {val_loss:.4f}, Val F1: {val_f1:.4f}")

        # Update learning rate based on F1 score
        scheduler.step(val_f1)

        # Log to wandb
        wandb.log({
            "epoch": epoch + 1,
            "train_loss": train_loss,
            "train_f1": train_f1,
            "val_loss": val_loss,
            "val_f1": val_f1,
            "learning_rate": optimizer.param_groups[0]['lr']
        })

        # Save the best model
        if val_f1 > best_f1:
            best_f1 = val_f1
            torch.save(model.state_dict(), model_save_path)
            print(f"New best model saved with F1: {best_f1:.4f}")

            # Log best model to wandb
            wandb.run.summary["best_val_f1"] = best_f1
            wandb.run.summary["best_model_epoch"] = epoch + 1

    return history, best_f1

# Evaluation function with wandb logging
def evaluate_model(model, test_loader, emotion_cols):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    model.eval()

    all_preds = []
    all_targets = []

    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)

            outputs = model(input_ids, attention_mask, token_type_ids)

            # Apply sigmoid to get probabilities and threshold
            preds = (torch.sigmoid(outputs) > 0.5).cpu().detach().numpy()
            all_preds.append(preds)

            if 'targets' in batch:
                targets = batch['targets'].cpu().numpy()
                all_targets.append(targets)

    all_preds = np.vstack(all_preds)
    # If we have targets (for validation), calculate metrics
    if all_targets:
        all_targets = np.vstack(all_targets)

        # Calculate precision, recall, and F1 for each emotion
        results = {}
        f1_scores = []
        emotion_metrics = {}

        for i, emotion in enumerate(emotion_cols):
            # Skip emotions with no positive examples
            if np.sum(all_targets[:, i]) > 0:
                precision = precision_score(all_targets[:, i], all_preds[:, i], zero_division=0)
                recall = recall_score(all_targets[:, i], all_preds[:, i], zero_division=0)
                f1 = f1_score(all_targets[:, i], all_preds[:, i], zero_division=0)

                results[emotion] = {
                    'precision': precision,
                    'recall': recall,
                    'f1': f1
                }

                # Add to metrics for wandb logging
                emotion_metrics[f"{emotion}_precision"] = precision
                emotion_metrics[f"{emotion}_recall"] = recall
                emotion_metrics[f"{emotion}_f1"] = f1

                f1_scores.append(f1)

        # Calculate macro-average F1 score
        macro_f1 = np.mean(f1_scores)
        print(f"Macro-average F1 Score: {macro_f1:.4f}")

        # Log all emotion metrics to wandb at once
        emotion_metrics["macro_f1"] = macro_f1
        wandb.log(emotion_metrics)

        # Print detailed results for each emotion
        for emotion, metrics in results.items():
            print(f"{emotion}: Precision={metrics['precision']:.4f}, Recall={metrics['recall']:.4f}, F1={metrics['f1']:.4f}")

        return all_preds, macro_f1, results

    return all_preds

# Main experiment function with wandb integration
def run_experiment():
    # Define file paths with clear print statements for debugging
    train_path = '/content/drive/MyDrive/train.csv'
    test_path = '/content/drive/MyDrive/test.csv'
    model_save_path = '/content/drive/MyDrive/best_albert_model.pt'
    predictions_path = '/content/drive/MyDrive/albert_predictions.csv'

    print(f"Training data path: {train_path}")
    print(f"Testing data path: {test_path}")
    print(f"Model save path: {model_save_path}")
    print(f"Predictions save path: {predictions_path}")

    # Check if files exist
    print(f"Training file exists: {os.path.exists(train_path)}")
    print(f"Testing file exists: {os.path.exists(test_path)}")

    # Initialize wandb run with project name and configuration
    wandb.init(
        project="emotion-classification",
        config={
            "model_type": "albert-base-v2",
            "max_length": 128,
            "batch_size": 16,
            "learning_rate": 2e-5,
            "epochs": 15,
            "dropout_rate": 0.3,
            "optimizer": "AdamW",
            "scheduler": "ReduceLROnPlateau"
        }
    )

    # Load datasets
    print(f"Loading training data from {train_path}")
    train_df = pd.read_csv(train_path)
    print(f"Loading testing data from {test_path}")
    test_df = pd.read_csv(test_path)

    # Print dataset info
    print(f"Training dataset shape: {train_df.shape}")
    print(f"Testing dataset shape: {test_df.shape}")

    # Define emotion columns
    emotion_cols = ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'love',
                   'optimism', 'pessimism', 'sadness', 'surprise', 'trust']

    # Check if emotion columns exist in the dataframes
    for col in emotion_cols:
        if col not in train_df.columns:
            print(f"Warning: {col} not found in training data")

    # Check for 'NONE' values in test data before conversion
    print("Checking for 'NONE' values in test data:")
    for col in emotion_cols:
        if col in test_df.columns:
            unique_vals = test_df[col].unique()
            if 'NONE' in unique_vals:
                print(f"Column {col} contains 'NONE' values")

    # Handle 'NONE' values and convert emotion columns to numeric - improved method
    for col in emotion_cols:
        if col in train_df.columns:
            # Replace 'NONE' with NaN first (case-insensitive)
            train_df[col] = train_df[col].replace(['NONE', 'None', 'none'], np.nan)
            # Then convert to float using to_numeric with coercion
            train_df[col] = pd.to_numeric(train_df[col], errors='coerce').fillna(0).astype(np.float32)

        if col in test_df.columns:
            # Replace 'NONE' with NaN first (case-insensitive)
            test_df[col] = test_df[col].replace(['NONE', 'None', 'none'], np.nan)
            # Then convert to float using to_numeric with coercion
            test_df[col] = pd.to_numeric(test_df[col], errors='coerce').fillna(0).astype(np.float32)

    # Check class distribution
    print("Emotion distribution in training set:")
    emotions_dist = train_df[emotion_cols].sum().sort_values(ascending=False)
    print(emotions_dist)

    # Log class distribution to wandb
    for emotion in emotion_cols:
        wandb.run.summary[f"{emotion}_count"] = float(train_df[emotion].sum())

    # Initialize tokenizer
    tokenizer = AlbertTokenizer.from_pretrained('albert-base-v2')

    # Create datasets
    train_dataset = EmotionDataset(train_df, tokenizer)
    test_dataset = EmotionDataset(test_df, tokenizer)

    # Split training data into train and validation sets (90/10 split)
    train_size = int(0.9 * len(train_dataset))
    val_size = len(train_dataset) - train_size
    # Use a fixed generator for reproducibility
    generator = torch.Generator().manual_seed(42)
    train_subset, val_subset = torch.utils.data.random_split(
        train_dataset, [train_size, val_size], generator=generator
    )

    # Create data loaders with a smaller batch size if needed
    batch_size = 16
    train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_subset, batch_size=batch_size)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)

    # Calculate class weights for handling imbalance
    class_weights = []
    total_samples = len(train_df)

    for col in emotion_cols:
        pos_samples = train_df[col].sum()
        # Avoid division by zero
        if pos_samples == 0:
            class_weights.append(1.0)
        else:
            # More weight to minority class
            class_weights.append(total_samples / (2 * pos_samples))
    class_weights = torch.tensor(class_weights, dtype=torch.float)
    print("Class weights:", class_weights)

    # Log class weights to wandb
    for i, emotion in enumerate(emotion_cols):
        wandb.run.summary[f"{emotion}_weight"] = class_weights[i].item()

    # Initialize model with a slightly higher dropout for regularization
    model = EmotionClassifier(n_classes=len(emotion_cols), dropout_rate=0.3)

    # Watch the model with wandb to track gradients and parameters
    wandb.watch(model, log="all")

    # Train model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    class_weights = class_weights.to(device)
    criterion = nn.BCEWithLogitsLoss(pos_weight=class_weights)

    # Train model
    history, best_f1 = train_model(model, train_loader, val_loader, epochs=15, lr=2e-5, model_save_path=model_save_path)

    # Log best model to wandb
    wandb.log({"best_f1": best_f1})

    # Load best model
    print(f"Loading best model from {model_save_path}")
    model.load_state_dict(torch.load(model_save_path))

    # Evaluate on test set
    if all(col in test_df.columns for col in emotion_cols):
        predictions, macro_f1, detailed_results = evaluate_model(model, test_loader, emotion_cols)

        # Plot F1 scores for each emotion
        emotions = list(detailed_results.keys())
        f1_scores = [detailed_results[emotion]['f1'] for emotion in emotions]

        plt.figure(figsize=(12, 6))
        sns.barplot(x=emotions, y=f1_scores)
        plt.title('F1 Score by Emotion')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig('/content/drive/MyDrive/emotion_f1_scores_albert.png')
        plt.show()

        # Log the plot to wandb
        wandb.log({"f1_score_by_emotion": wandb.Image('/content/drive/MyDrive/emotion_f1_scores_albert.png')})
    else:
        # If test set doesn't have labels, just make predictions
        print("Test set doesn't have emotion labels, making predictions...")
        predictions = evaluate_model(model, test_loader, emotion_cols)

    # Create submission dataframe with ID and Tweet
    print("Creating prediction dataframe...")
    submission_df = test_df[['ID', 'Tweet']].copy()
    for i, emotion in enumerate(emotion_cols):
        submission_df[emotion] = predictions[:, i]

    # Save predictions to multiple locations for redundancy
    try:
        # 1. Save to Google Drive
        print(f"Saving predictions to {predictions_path}...")
        submission_df.to_csv(predictions_path, index=False)
        print(f"Predictions successfully saved to {predictions_path}")

        # 2. Save to local Colab storage
        local_path = '/content/albert_predictions.csv'
        submission_df.to_csv(local_path, index=False)
        print(f"Predictions saved to local Colab storage: {local_path}")

        # 3. Save to a new folder in Drive for better organization
        output_dir = '/content/drive/MyDrive/colab_outputs'
        os.makedirs(output_dir, exist_ok=True)
        output_path = f"{output_dir}/albert_predictions.csv"
        submission_df.to_csv(output_path, index=False)
        print(f"Predictions saved to organized folder: {output_path}")

        # Verify files were created
        for path in [predictions_path, local_path, output_path]:
            if os.path.exists(path):
                print(f"✓ File exists at {path}")
                print(f" File size: {os.path.getsize(path)} bytes")
            else:
                print(f"✗ File does not exist at {path}")

        # 4. Download the file directly to your local machine
        from google.colab import files
        files.download(local_path)
        print("File download initiated. Check your browser's download folder.")

    except Exception as e:
        print(f"Error saving predictions: {str(e)}")

        # Try saving to Colab local storage as backup
        try:
            local_path = '/content/albert_predictions.csv'
            submission_df.to_csv(local_path, index=False)
            print(f"Predictions saved to local Colab storage: {local_path}")

            # Download the file directly to your local machine
            from google.colab import files
            files.download(local_path)
            print("File download initiated. Check your browser's download folder.")
        except Exception as download_error:
            print(f"Error downloading file: {str(download_error)}")
            print("Please try downloading manually using the file browser.")

    # Plot training history
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Loss over epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['val_f1'], label='Validation F1')
    plt.title('F1 Score over epochs')
    plt.xlabel('Epoch')
    plt.ylabel('F1 Score')
    plt.legend()

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/training_plots_albert.png')
    plt.show()

    # Log the plot to wandb
    wandb.log({"training_plots": wandb.Image('/content/drive/MyDrive/training_plots_albert.png')})

    print("Training plots saved to Google Drive")

    # Also download the training plots
    try:
        from google.colab import files
        files.download('/content/drive/MyDrive/training_plots_albert.png')
        print("Training plots download initiated.")
    except Exception as e:
        print(f"Error downloading training plots: {str(e)}")

    # Finish the wandb run
    wandb.finish()

if __name__ == "__main__":
    run_experiment()
```
